## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io
import os
```

##  Loading image

```python
image_path='C:\\Users\\KIIT\\Downloads\\cat.jpg'
original_img = io.imread(image_path)
```

## Displaying the original image

```python
plt.imshow(original_img)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

## Reshaping the image

```python
X_img = np.reshape(original_img, (original_img.shape[0] * original_img.shape[1], 3))
```

## Running K-Means Clustering

```python
K = 16  # Number of clusters(colors)
kmeans = KMeans(n_clusters=K, random_state=0).fit(X_img)

# Get the centroids and the labels (cluster assignments for each pixel)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

## Replacing Each Pixel Value with its Cluster Centroid

```python
compressed_img = centroids[labels].astype(np.uint8)
```

## Reshaping the Image Back to Its Original Shape

```python
compressed_img = np.reshape(compressed_img, (original_img.shape[0], original_img.shape[1], 3))
```

## Displaying the compressed image

```python
plt.imshow(compressed_img)
plt.title(f'Compressed Image with {K} colors')
plt.axis('off')
plt.show()

```

```python
# Saving the compressed image
compressed_image_path = r'C:\Users\KIIT\Desktop\compressed_image.jpg'
io.imsave(compressed_image_path, compressed_img)
```
